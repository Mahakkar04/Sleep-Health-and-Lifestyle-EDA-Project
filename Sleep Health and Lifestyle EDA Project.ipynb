# Data Cleaning-------------------------------------------

The dataset was cleaned to ensure accuracy, consistency, and reliability for analysis. Missing values in numerical features were handled using median imputation, while categorical variables were filled using appropriate methods such as mode or "Unknown" categories. Duplicate records were checked and removed where necessary. Data types were verified and corrected to maintain consistency across all columns. These preprocessing steps improved the overall quality of the dataset and prepared it for meaningful exploratory data analysis.


```python
# Check missing values in each column and sort them in descending order
# to quickly identify columns with the highest number of null values.
sleep_df.isnull().sum().sort_values(ascending=False)
```

```python
# Replace missing values in the 'Sleep Disorder' column with 'Unknown'.
# Missing values are treated as unavailable information rather than
# assuming the absence of a sleep disorder.

sleep_df['Sleep Disorder'].fillna('Unknown', inplace=True)
```

```python
# Fill missing values in numerical columns using the median.
# Median is used because it is robust to outliers and
# better represents the central value of the data.

num_cols = [  'Age', 'Sleep Duration', 'Quality of Sleep',
              'Physical Activity Level', 'Stress Level',
              'Heart Rate', 'Daily Steps',
              'Water_Intake_L', 'Screen_Time_Hours']

for col in num_cols:
    sleep_df[col].fillna(sleep_df[col].median(), inplace=True)
```

```python
# Replace missing values in BMI_Category with the most frequent category (mode),
# as BMI_Category is a categorical feature and mode provides a suitable estimate.

sleep_df['BMI_Category'].fillna(sleep_df['BMI_Category'].mode()[0],inplace=True)
```

```python
# Replace missing values in Alcohol_Consumption with the most frequent category (mode),
# as it is a categorical feature and mode helps preserve the overall distribution of the data.

sleep_df['Alcohol_Consumption'].fillna(sleep_df['Alcohol_Consumption'].mode()[0],inplace=True)
```

```python
# Remove the 'BMI Category' column to simplify the dataset by keeping a single BMI-related feature
sleep_df.drop(columns=['BMI Category'], inplace=True)
```

# Exploratory Data Analysis (EDA) and Visualization-------------------------------

#### Q1.Which occupation has the highest average sleep score?

```python
# Calculate average sleep score by occupation
avg_sleep_score = sleep_df.groupby('Occupation')['Sleep_Score'].mean().round(2).sort_values(ascending=False)
print(avg_sleep_score)
```

```python
avg_sleep_score = sleep_df.groupby('Occupation')['Sleep_Score'].mean().round(2).sort_values()

avg_sleep_score.plot(
    kind='barh',
    figsize=(8,5),
    color='#1F4E79',
    edgecolor='black'
)

plt.title('Average Sleep Score by Occupation')
plt.xlabel('Average Sleep Score')
plt.ylabel('Occupation')
plt.tight_layout()
plt.show()
```

**Observation**
Scientists recorded the highest average sleep score (77.38), while Sales Representatives had the lowest (72.05). The difference of 5.33 points indicates noticeable variation in sleep health across occupations.

**Insight**
Occupations such as Scientist, Manager, and Nurse show better sleep outcomes, with average sleep scores above 74.9. In contrast, Salesperson and Sales Representative roles exhibit comparatively lower sleep scores, suggesting potential impacts from work-related stress or irregular schedules.

**Recommendation**
Organizations should promote healthy sleep practices and work-life balance, particularly for sales-related roles with sleep scores below 73. Studying the habits of higher-scoring occupations like Scientists (77.38) may help improve employee well-being across other professions.

The correlations among Heart Rate, Resting HR, Oxygen Saturation, and Sleep Score are extremely weak, ranging from -0.02 to 0.02. **Insight**
Since all correlation values are very close to zero, these health metrics show little to no linear relationship with one another in this dataset. Changes in one metric do not appear to strongly influence the others.

**Recommendation**
